It seems there may be duplicate data in the Google data. Data may have been obtained at different times leading to duplication of data with slightly different values(for example in rating count in more popular apps)

```python
#Checking for duplicate data using a popular app
for app in googleplay:
    name= app[0]
    if name == 'Facebook':
        print(app)
```

```python
duplicate_apps =[]
unique_apps =[]

for app in googleplay:
    name = app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
print('Number of duplicated apps:', len(duplicate_apps))
print('Examples include:', duplicate_apps[:11])
```

There are 1181 entries of duplicated apps, to solve this problem we will remove the duplicated rows leaving only one. As seen with "Facebook" above, the difference between the rows was the review count. The app with the highest review count is likely to be the most recent, as such, we will use that as our criterion to remove duplicates.

```python
reviews_max ={}
for app in googleplay:
    name=app[0]
    n_reviews= float(app[3])
    if name in reviews_max and reviews_max[name]<n_reviews:
        reviews_max[name]=n_reviews
    elif name not in reviews_max:
        reviews_max[name]=n_reviews
```

```python
print('Expected number of rows: ', len(googleplay)-1181)
print('New number of rows: ', len(reviews_max))
#Excluding headers
```

```python
android_clean=[]
already_added=[]
for app in googleplay:
    name=app[0]
    n_reviews=float(app[3])
    if (reviews_max[name]==n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

Checking for and removing non-English language apps

```python
def non_eng(string):
    for cha in string:
        if ord(cha)>127:
            return False
    return True
```

As seen in the tests, "False" was displayed for the non-English characters "„ÅÇ„ÅÑ" but the app with an emoji in the title was labeled as non-English as emojis fall outside of the ASCII range for English(0-127). To minimize the potential loss of apps to this, we only remove apps that have more than 3 characters outside of the ASCII range.

```python
def is_eng(string):
    non_eng=0
    for cha in string:
        if ord(cha)>127:
            non_eng+=1
    if non_eng>3:
        return False
    else:
        return True
```

Our validation strategy is to build a minimal version of the app for Google Play, develop it further if there is good user response, then if it is profitable after 6 months, build an iOS version to add to the App Store. As such, we must find genre profiles that suit both app stores and customers.

The columns we will take most interest in describe the genre of the app, for the App Store being "Prime Genre" and for android being "Genres" and "Category", and columns detailing ratings/reviews and review counts.

```python

def freq_table(dataset, index):
    table={}
    total=0
    for app in dataset:
        total+=1
        value=app[index]
        if value in table:
            table[value]+=1
        else:
            table[value]=1
    table_percent={}
    for key in table:
        pct=(table[key]/total)*100
        table_percent[key]=pct
    return table_percent
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

## Observing Apps by Genre on the Google Play Store

There is a column showing number of installs in the data so we can simply pull from that. The numbers are not precise but the information is still very useful.

```python
display_table(free_android, 5)
```

We need to convert each install number to float ‚Äî this means that we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error. 

We'll do this directly in the loop below, where we also compute the average number of installs for each genre (category).

```python
categories_android = freq_table(free_android, 1)

for category in categories_android:
    total = 0
    len_category = 0
    for app in free_android:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```

From the list, we can see that "Communication" apps have the most installs with over 38 million installs, however this might be skewed by some giants.

```python
for app in free_android:
    if app[1] == 'COMMUNICATION' and (app[5] == '1,000,000,000+' or app[5] == '500,000,000+' or app[5] == '100,000,000+'):
        print(app[0], ':', app[5])
```

If we removed all the communication apps that have over 100 million installs, the giants, the average would be greatly reduced. Still a high number, but vastly different from the perception we had from looking at the surface data.

```python
under_100_m = []

for app in free_android:
    n_installs = app[5]
    n_installs = n_installs.replace(',', '')
    n_installs = n_installs.replace('+', '')
    if (app[1] == 'COMMUNICATION') and (float(n_installs) < 100000000):
        under_100_m.append(float(n_installs))
        
sum(under_100_m) / len(under_100_m)
```
